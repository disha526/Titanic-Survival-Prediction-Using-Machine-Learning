#MATPLOTLIB Tutorial


```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Use NumPy to genrate random data """
import numpy as np

x = np.linspace(0, 10, 25)
y = x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2).reshape(2,25))
```

```python
"""It takes 1 command to draw"""
pylab.plot(x,y,'r')
```

```python
"""Drawing a subgraph"""
pylab.subplot(1,2,1) #(rows,columns,indexes)
pylab.plot(x,y,'r--') #The third parameter here determines color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```
