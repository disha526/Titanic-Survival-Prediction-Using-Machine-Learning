# Create a new column Result (Pass if Marks >= 50 else Fail)
#Get all students who failed
# Find average Marks city-wise
# Find top 5 students based on Marks
# Create new CSV file of students with Marks > 85
# Find students whose Attendance > 90 AND Marks > 85
# Replace course "BCA" with "Computer Science"
# Save updated DataFrame into a new CSV file

```python
f_students = df[df['Result'] == 'Fail']
display(f_students)
```

```python
avg_marks_city= df.groupby('City')['Marks'].mean()
display(avg_marks_city)
```

```python
df['Course'] = df['Course'].replace('BCA', 'Computer Science')
display(df.head())
```

```python
high_attend_high_marks = df[(df['Attendance'] > 90) & (df['Marks'] > 85)]
display(high_attend_high_marks)
```

```python
df[df['Marks'] > 85].to_csv('top_students.csv', index=False)
```
